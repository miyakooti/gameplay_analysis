### ただしこのままでは2値データの数が多すぎるため、principal_pointsを用いて集約することを考えます。

```python
# 2値データだけ抽出
df_for_principal = df[['for_fun', 'male', 'has_work', 'is_single_player', 'is_highly_educated']]
```

```python
df_for_principal
```

```python
lim = 1000
data = df_for_principal.to_numpy()
data = data[:lim, :]
```

### 機械の力に頼り全探索をし、最も距離が小さくなるような分布の代表点を見つけます。

```python
principal_points = None

###STEP1データをインポートする###

(n,p)=np.shape(data)

###STEP1 Prinsipal Pointsの候補ベクトルを作成（ちょっとおしゃれな事やっています）###
pmatrix=np.empty((2**p,p))
for i in range(2**p):
    text = f"0{p}b"
    pmatrix[i,]=(list(format(i, text)))#0の後に変数の数がくる(ex.4変数のとき04b)


#Step2　principal pointsを探索
sum2=1000
for i in range (2**p-2):
    for j in range ((i+1),2**p-1):
        for k in range((j+1),2**p):    
            sum1=0
            for num in range (n):
                sum1=sum1+min(p-np.sum(pmatrix[i]==data[num]),p-np.sum(pmatrix[j]==data[num]),p-np.sum(pmatrix[k]==data[num]))
            if sum1<=sum2:
                sum2=sum1
                pp1=i
                pp2=j
                pp3=k
                print('距離',sum2)
                print(list(format(pp1, text)) ,list(format(pp2, text)),list(format(pp3, text)))
                
                principal_points = [list(format(pp1, text)), list(format(pp2, text)), list(format(pp3, text))]
                
print(principal_points)
```

### これでprincipal_pointがもとまりました。

```python
principal_points = [[int(value) for value in sublist] for sublist in principal_points]
principal_points
```

### 2値データのスキーマを参考にすると、
# 1個目のpp:「単なる男性」
# 2個目のpp:「仕事を持っている高学歴な男性」
# 3個目のpp:「ゲームは楽しみのためにしかやらない男性」
### と解釈することができます。

### これをもとに、一番近いppに分類し、分類結果をダミー変数化します。

```python
# 新しい特徴量の初期化
principal_features = np.zeros((len(df), 3), dtype=int)

# 各行に対して最も近いプリンシパルポイントを特定
for i, row in enumerate(df.iterrows()):
    distances = np.sum(np.abs(np.array(row[1][['for_fun', 'male', 'has_work', 'is_single_player', 'is_highly_educated']]) - np.array(principal_points)), axis=1)
    nearest_point_index = np.argmin(distances)
    principal_features[i, nearest_point_index] = 1

# 新しい特徴量をDataFrameに追加
df['principal_1'] = principal_features[:, 0]
df['principal_2'] = principal_features[:, 1]
df['principal_3'] = principal_features[:, 2]
```

# 3：分析モデルの定式化

誤差ベクトル𝜺に確率変数を導入した、重回帰分析を利用。Xは確率変数でないとしている。


# y=Xβ+ε
# ε~N(0,σ^2 I)


# 4：結果と考察

```python
import statsmodels.api as sm

# 説明変数と目的変数のデータを準備
X = df.drop(['SWL_T'], axis=1)
y = df['SWL_T']

# 標準化？でも変数間で比べるわけじゃないからイイや
# scaler = MinMaxScaler()
# X = scaler.fit_transform(X)

# 定数項を追加
X = sm.add_constant(X)

# 重回帰分析モデルの構築と学習
model = sm.OLS(y, X)
result = model.fit()

# 重回帰分析結果の表示
print(result.summary())

# AICの算出
aic = result.aic
print("AIC:", aic)
```

決定係数は0.199、AICは83473とそれほど精度の良いモデルは得られなかった。ただし集約したカテゴリ刈るデータであるprincipal_1,2,3のt値の絶対値は、他の変数と比べても大きくなっており、人生の満足度に大きく有意に関連していることがわかる。その中でも特に「ゲームは楽しみのためにしかやらない男性」という自堕落な特徴量が寄与していることがわかり、これは直感からでも納得がいく。その次に「単なる男性」、「仕事を持っている高学歴な男性」という順番になっており、自分の予想であった「仕事を持っている高学歴な男性は、人生の満足度が高いはずだ」という仮説を一部支持する結果となった。このような男性よりも「ゲームを楽しみでしかやらない男性」の方が満足度が高いのは、変に高いプライドを持っておらず満足度のハードルが低いためであると考えることができる。


t値の高さから、やはりprincipal_pointsで分類することで、より深みがあり意味のある特徴量が作成できているということが考察できる。

# 結論、今後の課題

満足度は「男性という性質そのもの」と、「男性の中でも特定の習慣を持っていること」が重要となることがわかった。そのため世の中のゲーマーの男性は、ゲームを成長や勝利のためにプレイするというよりは、ただただ楽しみのためだけにやるということが、人生の満足度につながる。仕事や学歴も重要であるが、もっと重要なのは子供心を忘れずに、より純粋にゲームを楽しむことなのである。

また課題として、AICのさらなる低下や決定係数の上昇が挙げられる。そのために2値型変数の作り方や条件文の変更、principal_pointsのk値の変更などを試して、より良い数字が出ればそちらを選択するのが良いだろう。
